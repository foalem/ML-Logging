```python
def filter_dataframe(df):
    filtered_df = df[(df['number_contributor'] > 1) & (df['number_stars'] > 1) & (df['number_commit'] >= 100)]
    return filtered_df
```

```python
data['number_contributor'] = pd.to_numeric(data['number_contributor'], errors='coerce')
```

```python
data.dropna(subset=['number_contributor'], inplace=True)
```

```python
data['number_stars'] = pd.to_numeric(data['number_stars'], errors='coerce')
data.dropna(subset=['number_stars'], inplace=True)
```

```python
def plot_violin_boxplot(df, filename):
    plt.rcParams.update({'font.size': 12})
    # Create a bar plot of the logging types
    plt.figure(dpi=500)
    colors = sns.color_palette("husl", len(df.columns))
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))
    handles = []
    labels = []
    for i, col in enumerate(df.columns):
        ax = axes[i//2, i%2]
        q1, q3 = np.percentile(df[col], [25, 75])
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        df_filtered = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
        max_val = df_filtered[col].max()
        ax.set_ylim(0, max_val)
        sns.set(font_scale=1.2)
        violin = sns.violinplot(data=df_filtered[col], ax=ax, color=colors[i], alpha=0.8, grid=False)
        box = sns.boxplot(data=df_filtered[col], ax=ax, color=colors[i], width=.15, boxprops=dict(facecolor=colors[i], alpha=0.8), medianprops=dict(color='white'))
        handles.append(violin.collections[0])
        labels.append(col)
        ax.set_xlabel(col, fontsize=16)
        ax.set_xticklabels(ax.get_xticks().astype(int), fontsize=12)
#         ax.set_ylabel(col, fontsize=16)
        ax.tick_params(axis='both', which='major', labelsize=12)
    fig.legend(handles, labels, loc='lower left', bbox_to_anchor=(0.1, 1.05), title='Metrics', fontsize=12)
    fig.supxlabel("Project's Metrics", fontsize=20)
    fig.subplots_adjust(hspace=0.5)
    plt.tight_layout()
    plt.savefig(filename, format='pdf', dpi=300,bbox_inches='tight')
    plt.show()
```

```python
plot_violin_boxplot(data_project[["Stars", "Contributors", "Commits", "Size"]], 'Project_overview.pdf')
```
